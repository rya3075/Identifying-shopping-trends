```python
#Overall distribution of customer ages across the dataset
shop['Age'].value_counts()
```

```python
shop['Age'].mean()
```

```python
shop['Age_Category'] = pd.cut(shop['Age'], bins = [0, 15,18, 30, 50, 70], labels = ['Child', 'Teen', 'Young_Adult', 'Middle_Aged',
                                                                                    'Old'])
```

```python
fig = px.histogram(shop, x = 'Age_Category', title = 'Distribution of Customer Age Categories')
fig.show()
```

```python
#Affect of promo codes on customer spending
shop_group = shop.groupby('Promo Code Used')['Purchase Amount (USD)'].mean().reset_index()
shop_group
```

```python
fig = px.sunburst(shop, path = ['Gender','Promo Code Used'], values = 'Purchase Amount (USD)', title = 'Promo Code Usage and Subscription Status')
fig.show()
```

```python
fig = px.bar(shop_group, x = 'Promo Code Used', y = 'Purchase Amount (USD)', title = 'Average Purchase Amount by Promo Code Usage')
fig.show()
```
